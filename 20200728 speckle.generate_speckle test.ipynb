# Generated simulated speckle

In this notebook, we show how to generate a speckle pattern using some black magic code.

```python
import supreme as sp
# awojdyla@lbl.gov, July 2020

# propagation distance
z_m = 100e-3
# pixel size
dx_m = 1e-6
# screen size
Dx_m = 2e-3-dx_m

img = sp.generate_speckle(dx_m, Dx_m, z_m)

import matplotlib.pyplot as plt

extent = (-Dx_m/2 * 1e6, +Dx_m/2 * 1e6, -Dx_m/2 * 1e6, +Dx_m/2 * 1e6)
fig=plt.figure(figsize=(5, 5), dpi= 120, facecolor='w', edgecolor='k')
plt.imshow(img, extent=extent, cmap = 'gray')
plt.xlabel('Horizontal Position [um]')
plt.ylabel('Vertical Positon [um]')
plt.show()
```
