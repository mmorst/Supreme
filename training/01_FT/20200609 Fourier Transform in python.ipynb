# Fourier Transform in python

We are going to explore how to perform a Fourier Transform in python and use it for signal analysis

```python
#awojdyla@lbl.gov, June 2020
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create time signal

```python
x_s = np.linspace(0, 10,1001)
```

```python
f0_cpm = 2
y = np.sin(2*np.pi*f0_cpm*x_s)
```

```python
plt.plot(x_s, y)
plt.show()
```

```python
Y = np.fft.fft(y)
plt.plot(abs(Y)) #no units!
plt.title('A spectrum fresh from the oven!')
plt.show()
```

```python
Yc = np.fft.fftshift(Y)
plt.plot(abs(Yc)) #no units!
plt.title('A zero-centered spectrum')
plt.show()
```

```python
# frequency scale
real_axis = x_s
fs = 1/(real_axis[1]-real_axis[0])
Nfft=len(real_axis)
            
df = fs/Nfft
f_cpm = np.linspace(0,(fs-df),Nfft) - (fs-np.mod(Nfft,2)*df)/2
```

```python
plt.plot(f_cpm, abs(Yc)) #no units!
plt.title('A scaled spectrum')
plt.show()
```

```python
plt.plot(f_cpm, abs(Yc)) #no units!
plt.xlabel('frequency [s^-1]')
plt.ylabel('FT magnitude')
plt.title('A properly plotted spectrum')
plt.xlim((-5,5))

plt.show()
```
