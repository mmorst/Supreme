```python
import numpy as np # import numpy library as np
import matplotlib.pyplot as plt # import matplotlib library as plt
x_s = np.linspace(0,10,1001) # define the x-axis as an array
f0 = 1 # giving a value fot the frequency
y = np.sin(2*np.pi*f0*x_s) # defining a sine wave, note: this array is the same length as x_s
plt.plot(x_s, y) # plots x_s as a function of y
plt.show() # display a graph of the plot
```

```python
fz_cpm = .2
z = np.sin(2*np.pi*fz_cpm*x_s**2)
plt.plot(x_s, np.abs(z)) 
plt.show()
```
