```python
import numpy as np
from PIL import Image
from skimage.transform import warp, warp_polar
import matplotlib.pyplot as plt

img = Image.open(r'C:\\Users\User\\Desktop\\Supreme\\Super\\supreme-00-01.png', "r")
data = np.asarray(img)

S = np.fft.fftshift(np.fft.fft2(data))
S01 = abs(S)**0.1
plt.imshow(S01)
plt.show()

plt.imshow(warp_polar(S01))
plt.show()

```

```python
s_warp = warp_polar(np.abs(S)**2)
circ_sum = np.sum(s_warp, 0)
plt.plot(circ_sum)
plt.ylim((0, 1000000000000))
plt.xlim((0,500))
plt.show()
```
