```python
# Creating a sine wave:

import numpy as np # import numpy library as np
import matplotlib.pyplot as plt # import matplotlib library as plt
x_s = np.linspace(0,10,1001) # define the x-axis as an array
f0 = 1 # giving a value fot the frequency
y = np.sin(2*np.pi*f0*x_s) # defining a sine wave, note: this array is the same length as x_s
plt.plot(x_s, y) # plots x_s as a function of y
plt.show() # display a graph of the plot
```

```python
# frequency scale
real_axis = x_s
fs = 1/(real_axis[1]-real_axis[0])
Nfft=len(real_axis)
            
df = fs/Nfft
f_cpm = np.linspace(0,(fs-df),Nfft) - (fs-np.mod(Nfft,2)*df)/2
```

```python
Y = np.fft.fft(y) # this creat a fourier transformation of "y" and assign it a new variable "Y"
Yc = np.fft.fftshift(Y)
```

```python
# Fourier transform of a sine wave (frequency)
plt.plot(f_cpm, abs(Yc))
plt.xlabel('frequency [s^-1]')
plt.ylabel('FT mag')
plt.title('good plot')
plt.xlim((0,10))
plt.show()
```
